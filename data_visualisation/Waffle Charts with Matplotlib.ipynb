# Waffle Charts With Matplotlib

```python
from pywaffle import Waffle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lith_dict = {'LITH': ['Shale', 'Sandstone', 
                      'Sandstone/Shale', 'Chalk', 
                      'Limestone', 'Marl', 'Tuff'],
             'Well1': [61,15, 10, 5, 
                            5, 3, 1],
             'Well2': [35 ,21, 16, 12, 
                            7, 5, 4]}

lith_data_df = pd.DataFrame.from_dict(lith_dict)
```

```python
lith_data_df
```

```python
colours = ['#8dd3c7', '#deb887', '#bebada', '#fb8072', 
           '#80b1d3', '#fdb462', '#b3de69']
```

```python
plot_labels = [f'{i} ({str(j)} %)' for i,j in zip(lith_data_df.LITH, 
                                                    lith_data_df.Well1)]
```

```python
plot_labels
```

```python
plt.figure(FigureClass=Waffle, figsize=(10, 10), rows=5, columns=20,
          values=list(lith_data_df['Well1']),
          colors=colours, 
          labels=plot_labels,
          legend={'loc':'lower center', 'bbox_to_anchor':(0.5, -0.8), 'ncol': 3, 'fontsize':12})
plt.show()
```

```python
plt.figure(FigureClass=Waffle, figsize=(10, 10), rows=5, columns=20,
          values=list(lith_data_df['Well1']),
          colors=colours, 
          labels=plot_labels,
          legend={'loc':'lower center', 'bbox_to_anchor':(0.5, -0.8), 'ncol': 3, 'fontsize':12}, starting_location='NW')
plt.show()
```

```python
plt.figure(FigureClass=Waffle, figsize=(10, 10), rows=5, columns=20,
          values=list(lith_data_df['Well2']),
          colors=colours, 
          labels = [f'{i} ({str(j)} %)' for i,j in zip(lith_data_df.LITH, 
                                                    lith_data_df.Well2)],
          legend={'loc':'lower center', 'bbox_to_anchor':(0.5, -0.8), 'ncol': 3, 'fontsize':12}, starting_location='NW')
plt.show()
```

```python
plt.figure(FigureClass=Waffle, 
           plots= {211: {'values':list(lith_data_df['Well1']), 
                         'labels': [f'{i} ({str(j)} %)' for i,j in zip(lith_data_df.LITH, 
                                                    lith_data_df.Well1)],
                        'legend':{'loc':'center left', 'bbox_to_anchor':(1.0, 0.5), 'ncol': 1, 'fontsize':10},
                        'title': {'label':'Well 1 Lithology Composition', 'fontsize':18}
                        },
                  212: {'values':list(lith_data_df['Well2']), 
                         'labels': [f'{i} ({str(j)} %)' for i,j in zip(lith_data_df.LITH, 
                                                    lith_data_df.Well2)],
                        'legend':{'loc':'center left', 'bbox_to_anchor':(1.0, 0.5), 'ncol': 1, 'fontsize':10},
                        'title': {'label':'Well 2 Lithology Composition', 'fontsize':18}
                       }
                  },
           figsize=(15, 10), rows=5, columns=20,
          
          colors=colours, 
          starting_location='NW')
plt.show()
```

```python
off_colour = 'lightgrey'
```

```python
fig, axs = plt.subplots(len(lith_data_df), 1, figsize=(10, 15))

for (i, ax), color in zip(enumerate(axs.flatten()), colours):
    plot_colours = [color, off_colour]
    perc = lith_data_df.iloc[i]['Well1']
    values = [perc, (100-perc)]
    lith = lith_data_df.iloc[i]['LITH']
    Waffle.make_waffle(ax=ax, rows=5, columns=20, values=values, colors=plot_colours)
    ax.set_title(lith)

plt.tight_layout()
plt.show()
    
```
